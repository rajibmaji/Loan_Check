```python
data_new = data[pd.notnull(data['Years in current job'])]
```

```python
data_new.apply(lambda x: sum(x.isnull()))
```

```python
data_new['Credit Score'].fillna(value=data_new['Credit Score'].mean(),inplace=True)     
data1=data_new.round(decimals=2)
```

```python
data1['Annual Income'].fillna(value=data1['Annual Income'].mean(),inplace=True)     
data2=data1.round(decimals=2)
```

```python
data2['Months since last delinquent'].fillna(value=data2['Months since last delinquent'].mean(),inplace=True)     
data3=data2.round(decimals=2)
```

```python
data3.dropna(inplace=True)
```

```python
class_set= data3['Loan Status']
pd.unique(class_set)
```

```python
class_set= data3['Home Ownership']
pd.unique(class_set)
```

```python
data3['Home Ownership'] = data3['Home Ownership'].map({'Home Mortgage':0,'Own Home':1,'Rent':2,'HaveMortgage':3})
data3['Term'] = data3['Term'].map({'Short Term':0,'Long Term':1})
data3['Loan Status'] = data3['Loan Status'].map({'Fully Paid':0,'Charged Off':1})
data3.shape
```

```python
q=data3['Current Loan Amount'].quantile(0.99)
data4=data3[data3['Current Loan Amount']<q]
data4.shape
```

```python
print(data4.columns)
```

```python
X1=data4[['Current Loan Amount',
       'Credit Score', 'Annual Income',
        'Monthly Debt', 'Years of Credit History',
       'Months since last delinquent', 'Current Credit Balance',
       'Maximum Open Credit']]
Y=data4['Loan Status']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X1)
X=scaler.transform(X1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3, random_state=66,shuffle=False)
```

```python
Y_test.shape
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(Y_test,y_pred))
```

```python
print(accuracy_score(Y_test,y_pred))
```
